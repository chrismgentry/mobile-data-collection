```r
#google sheet to csv
drive_deauth()
drive_download(as_id("1KyNkvqJej969kKexz0_6vb_AZ8zp5r-tHE7CdPmFhmI"), overwrite = TRUE, type = "csv")
demo <- as.data.frame(read.csv("BIOL_5660_Demo (Responses).csv")) %>% drop_na(c("Longitude","Latitude"))
```

```r
#google sheet responses
head(demo)
```

```r
#leaflet - needs to be downloaded and opened in a browser
html <- demo %>%
      leaflet() %>%
      addTiles(group = "OSM")%>%
        addProviderTiles(providers$CartoDB.Positron, group = "CartoDB") %>%
        addProviderTiles(providers$Esri.NatGeoWorldMap, group = "NatGeo") %>%
        addProviderTiles(providers$Esri.WorldImagery, group = "ESRI") %>%
      addCircleMarkers(popup = demo$Species,
                       label = demo$DBH,
                       radius = demo$DBH,
                       color = "grey",
                       fillColor = "red",
                       fillOpacity = 0.7,
                       labelOptions = labelOptions(noHide = T, direction = "bottom",
                                                   offset=c(0,-20), textOnly = TRUE)) %>%
      addLayersControl(
        baseGroups = c("OSM", "CartoDB", "NatGeo", "ESRI"),
        options = layersControlOptions(collapsed = FALSE),
        overlayGroups = "Facilities") %>%
      addMiniMap(zoomLevelOffset = -4) %>%
      addScaleBar()
saveWidget(html, file = "biol_5660_demo.html")
```
